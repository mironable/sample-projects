```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data
titanic_df = pd.read_csv('data/train.csv')

# preview first five rows
titanic_df.head()
```

```python
# Questions to answer:

# 1 Passengers information (ages, gender, class, etc)
# 2 Where did the passengers come from
# 3 Who was alone and who was with family
# 4 What factor(s) were important for someone's survival (class, women/children first, etc)
# 5 How the actual class of passengers is related to their deck (higher, lower, etc)
```

```python
# Question 1: Passengers information (age, gender, class, etc)

# To get a better insight on whether passenger's age can be a factor (e.g. children)
# we define as children passengers < 16 and apply this function to a new column ['Person']

def age_sex_sep(passenger):
    age,sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
    
# Indicate whether Person is male, female or child 
# as specified in the function with 'age' and 'sex' information
titanic_df['Person'] = titanic_df[['Age', 'Sex']].apply(age_sex_sep, axis=1)

# Visualise male/female/child ratio per class
figure = sns.factorplot('Pclass', data=titanic_df, hue='Person', kind='count', palette='summer', size=5)
(figure.set_axis_labels('Classes','Counts'))
figure = plt.title('Sex ratio based on class')

# A lot more males and children in the third class
```

```python
# A simple age distribution
titanic_df['Age'].hist(bins=80)
plt.title('Age distribution')

# with a mean approx 30 
titanic_df['Age'].mean()
```

```python
# More than one plots by setting a 'figure' object
figure_person = sns.FacetGrid(titanic_df, hue='Person', size=3, aspect=4)
# map each kdeplot by 'Age' column 
figure_person.map(sns.kdeplot, 'Age', shade=True)
# set limits for 'Age' 0 to the max age
old = titanic_df['Age'].max()
figure_person.set(xlim=(0,old))
figure_person.set(ylim=(0.00,0.09))
figure_person.add_legend()
figure_person = plt.title('Distribution of people based on age')

# For 'child' it goes past the 16 mark we've set earlier since bandwidth is present for this kernel density plot

figure_class = sns.FacetGrid(titanic_df, hue='Pclass', size=3, aspect=4)
figure_class.map(sns.kdeplot, 'Age', shade=True)
figure_class.set(xlim=(0,old))
figure_class.set(ylim=(0.00,0.04))
figure_class.add_legend()
figure_class = plt.title('Distribution of classes based on age')

# Much smoother distribution for class 1 vs. classes 2 and 3 which are skewed towards 20-30 range
```

```python
# It can be useful to know if the deck of a passenger reveals a relation -
# since this information is contained in the 'Cabin' we will keep just 
# letters (A,B,C,D,E,F,G) and Nan values for the 'Cabin' column will be dropped.

all_decks = []
for deck in titanic_df['Cabin'].dropna():
    all_decks.append(deck[0])
    
# To plot counts of people in each deck we create new dataframe since 
# the length is now different and cannot append to existing dataframe
deck_df = pd.DataFrame(all_decks)
deck_df.columns = ['Deck']
deck_df = deck_df[deck_df.Deck != 'T'] # keep everything that is not 'T' in data 
figure = sns.factorplot('Deck', data=deck_df, order=['A','B','C','D','E','F','G'],
               kind='count', palette='cool_d', size=5)
(figure.set_axis_labels('Decks','Counts'))
figure = plt.title('Counts of people per deck')
```

```python
# Question 2: Passengers' origin

titanic_df.head()
# According to Kaggle S, C and Q in the 'Embarked' column 
# correspond to Southampton, Cherbourg and Queenstown, respectively.

# Similarly we can visualise counts of the town each passenger comes from based on their class
figure = sns.factorplot('Embarked', data=titanic_df, hue='Pclass', kind='count', palette='bwr_d', size=5)
(figure.set_xticklabels(['Southampton', 'Cherbourg', 'Queenstown'])
       .set_axis_labels('','Counts'))
figure = plt.title("City of origin based on passenger's class")

# Almost all passengers boarded in Queenstown were third-class: reveals the economics of 
# Queenstown in 1912, whereas passengers boarded in Cherbourg were a lot of first-class people.
```

```python
# Question 3: travelling alone or with family

# Define 'Alone' column based on information from 'SibSp' and 'Parch', 
# If 'Alone status' = 0 then no family on-board. We set the status of it based 
# on this condition and get a feel of the counts of people travelling alone.

pd.options.mode.chained_assignment = 'warn'  # default='warn'

titanic_df['Alone status'] = titanic_df['SibSp'] + titanic_df['Parch']
titanic_df['Alone status'].loc[titanic_df['Alone status'] >0] = 'with family'
titanic_df['Alone status'].loc[titanic_df['Alone status'] == 0] = 'alone'
figure = sns.factorplot('Alone status', data=titanic_df, kind='count', palette='Purples_r', size=5)
figure.set_axis_labels('Alone status','Counts')
figure = plt.title('Travelling with family/alone')
```

```python
# Question 4: factors for someone's survival

# 4.1 Given that third class had much more males than females, let's 
# find out if 'Pclass' or 'gender' are contributing factors

figure = sns.factorplot('Pclass','Survived', data=titanic_df, hue='Person', palette='Blues_d', size=5, aspect=1.2)
figure = plt.title("Survival rate based on a person's class")
# If we look at the tedency of 'male' it seems that being a male is not really favorable of survival
# particularly for second and third classes.
```

```python
# 4.2 It would be interesting to see if 'Pclass' indeed contributes in someone's survival,
# based on 'Age'. Since we are dealing with lots of age datapoints it makes sense to seperate 
# them in bins.

age_bins = [10, 20, 30, 40, 60, 80]
figure = sns.lmplot('Age', 'Survived', data=titanic_df, hue='Pclass', 
                 palette='Blues_d', x_bins=age_bins, size=5, aspect=1.2)
figure = plt.title("Survival rate based on a passenger's age and class")

# We notice the high standard deviation on the older people in the first class.
# However the survival rate is less favourable for people >40.
```

```python
# 4.3 Similarly we can see how gender and age relate with survival:

figure = sns.lmplot('Age', 'Survived', data=titanic_df, hue='Sex', palette='Blues_d', 
                    x_bins=age_bins, size=5, aspect=1.2)
figure = plt.title("Relation of age and gender for survival")

# It looks like there might be better chance of survival for an older female  
# compared to an older male.
```

```python
# 4.4 Let's see if being at a particular deck had an effect on the passengers survival rate:

# Get first letter in 'Cabin' 
def get_deck(cabin):
    for level in cabin:
        return level[0]
# Apply function in a new column
titanic_df['Deck'] = titanic_df['Cabin'].dropna().apply(get_deck)

figure = sns.factorplot('Deck', 'Survived', data=titanic_df, order=['A','B','C','D','E','F','G'],
               palette='Blues_d', size=5, aspect=1.2)
figure = plt.title("Survival rate based on passenger's deck")

# Looks like being on decks B, D or E slightly improve the chances of survival. 
# These decks might also reveal the location of lifeboats.
```

```python
# 4.5 Did having a family enable you to get on a lifeboat 
# since you had people supporting you or was it better being alone and 
# not worry about your family.

figure = sns.factorplot('Alone status', 'Survived', data=titanic_df, hue='Sex',
               palette='Blues_d', size=5, aspect=1.2)
figure = plt.title("Travelling with family/alone as a factor for survival")

# It looks like being a single female would bring you closer to survival, 
# whereas single males did not have much of a chance. 
# On the other hand, chances for females with family appear to drop and
# being a male with family might enable you to get on a lifeboat.
```
